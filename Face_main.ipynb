```python
from tkinter import *
from tkinter import ttk
import tkinter
from PIL import Image, ImageTk
from time import strftime
from datetime import datetime
import os

from ipynb.fs.full.student import StudentSystem
from ipynb.fs.full.training import Train
from ipynb.fs.full.FaceReconizer import FaceReconizerSystem
from ipynb.fs.full.Attendance import AttendanceSystem
from ipynb.fs.full.Developer import DeveloperSystem
from ipynb.fs.full.Help import HelpSystem
```

```python
class face_reconization_System:
    def __init__(self, root):
        self.root = root
        self.root.geometry('1370x790+0+0')
        self.root.title('Face Reconigation System')
        
        # First Image
        img =Image.open('img/university.jpg') 
        img = img.resize((450,130), Image.ANTIALIAS)
        self.photoimg = ImageTk.PhotoImage(img)
        
        f_lbl = Label(self.root, image = self.photoimg)
        
        f_lbl.place(x=0,y=0,width=450, height=130)
        
        
        # Second Image
        img1 =Image.open('img/facialrecognition.png') 
        img1 = img1.resize((450,130), Image.ANTIALIAS)
        self.photoimg1 = ImageTk.PhotoImage(img1)
        
        f_lbl = Label(self.root, image = self.photoimg1)
        
        f_lbl.place(x=450,y=0,width=450, height=130)
        
        
        # Third image
        img2 =Image.open('img/university.jpg') 
        img2 = img2.resize((450,130), Image.ANTIALIAS)
        self.photoimg2 = ImageTk.PhotoImage(img2)
        
        f_lbl = Label(self.root, image = self.photoimg2)
        
        f_lbl.place(x=900,y=0,width=480, height=130) 
        
        # Background image
        img3 =Image.open('img/milky-way.jpg') 
        img3 = img3.resize((1370,710), Image.ANTIALIAS)
        self.photoimg3 = ImageTk.PhotoImage(img3)
        
        bg_img = Label(self.root, image = self.photoimg3)
        
        bg_img.place(x=0,y=130,width=1370, height=710)         
        
        
        
        # Putting title
        title_lbl = Label(bg_img,text='FACE    RECONIZATION      ATTENDENCE     SYSTEM',font=('times new roman',20,'bold'),bg='red',fg='white')
        title_lbl.place(x=0, y=0, width=1370, height=45)
        
        def time():
            string=strftime("%H:%M:%S %p")
            lbl.config(text=string)
            lbl.after(1000,time)
        lbl = Label(title_lbl,font=('times new roman',15,'bold'),bg='red',fg='white')
        lbl.place(x=0, y=0, width=110, height=50)
        time()
            
        # Student buttom 
        img4 =Image.open('img/student.jpg') 
        img4 = img4.resize((200,200), Image.ANTIALIAS)
        self.photoimg4 = ImageTk.PhotoImage(img4)
        
        # Up First-buttom
        b1= Button(bg_img, image=self.photoimg4,command=self.student_details,  cursor='hand2')
        b1.place(x=120, y=70, width=220, height=220)
        
        b1_1= Button(bg_img,text='Student details',command=self.student_details, cursor='hand2',font=('times new roman',14,'bold'),bg='darkblue',fg='white')
        b1_1.place(x=120, y=250, width=220, height=40)
        
        # Detect Face
        img5 =Image.open('img/face_detector1.jpg') 
        img5 = img5.resize((200,200), Image.ANTIALIAS)
        self.photoimg5 = ImageTk.PhotoImage(img5)
        
        # Up Second-buttom
        b1= Button(bg_img, image=self.photoimg5, cursor='hand2',command=self.face_data)
        b1.place(x=450, y=70, width=220, height=220)
        
        b1_1= Button(bg_img,text='Face Detector',cursor='hand2',command=self.face_data,font=('times new roman',14,'bold'),bg='darkblue',fg='white')
        b1_1.place(x=450, y=250, width=220, height=40)
        
        
        # Attendance Buttom
        img6 =Image.open('img/smart-attendance.jpg') 
        img6 = img6.resize((200,200), Image.ANTIALIAS)
        self.photoimg6 = ImageTk.PhotoImage(img6)
        
        # Up Third-buttom
        b1= Button(bg_img, image=self.photoimg6, cursor='hand2',command=self.attendance_data)
        b1.place(x=750, y=70, width=220, height=220)
        
        b1_1= Button(bg_img,text='Attendance',cursor='hand2',command=self.attendance_data,font=('times new roman',14,'bold'),bg='darkblue',fg='white')
        b1_1.place(x=750, y=250, width=220, height=40)
        
        
        #  Help Desk
        img7 =Image.open('img/help.jpg') 
        img7 = img7.resize((200,200), Image.ANTIALIAS)
        self.photoimg7 = ImageTk.PhotoImage(img7)
        
        # Up fourth-buttom
        b1= Button(bg_img, image=self.photoimg7, cursor='hand2',command=self.help_data)
        b1.place(x=1100, y=70, width=220, height=220)
        
        b1_1= Button(bg_img,text='Help Desk',cursor='hand2',command=self.help_data,font=('times new roman',14,'bold'),bg='darkblue',fg='white')
        b1_1.place(x=1100, y=250, width=220, height=40)
        
        #  Train face
        img8 =Image.open('img/di.jpg') 
        img8 = img8.resize((200,200), Image.ANTIALIAS)
        self.photoimg8 = ImageTk.PhotoImage(img8)
        
        # Down First-buttom
        b1= Button(bg_img, image=self.photoimg8, cursor='hand2',command=self.train_data)
        b1.place(x=120, y=340, width=220, height=220)
        
        b1_1= Button(bg_img,text='Training',cursor='hand2',command=self.train_data,font=('times new roman',14,'bold'),bg='darkblue',fg='white')
        b1_1.place(x=120, y=520, width=220, height=40)
        
        
        # Photos Face
        img9 =Image.open('img/BestFacialRecognition.jpg') 
        img9 = img9.resize((200,200), Image.ANTIALIAS)
        self.photoimg9 = ImageTk.PhotoImage(img9)
        
        # Down Second-buttom
        b1= Button(bg_img, image=self.photoimg9, cursor='hand2',command=self.open_img)
        b1.place(x=450, y=340, width=220, height=220)
        
        b1_1= Button(bg_img,text='Photos',cursor='hand2',command=self.open_img,font=('times new roman',14,'bold'),bg='darkblue',fg='white')
        b1_1.place(x=450, y=520, width=220, height=40)
        
         #  Developer face Buttom
        img10 =Image.open('img/dev.jpg') 
        img10 = img10.resize((200,200), Image.ANTIALIAS)
        self.photoimg10 = ImageTk.PhotoImage(img10)
        
        # Down Third-buttom
        b1= Button(bg_img, image=self.photoimg10, cursor='hand2',command=self.developer_data)
        b1.place(x=750, y=340, width=220, height=220)
        
        b1_1= Button(bg_img,text='Developer',cursor='hand2',command=self.developer_data,font=('times new roman',14,'bold'),bg='darkblue',fg='white')
        b1_1.place(x=750, y=520, width=220, height=40)
        
        
        # Exit face Desk
        img11 =Image.open('img/exit.jpg') 
        img11 = img11.resize((200,200), Image.ANTIALIAS)
        self.photoimg11 = ImageTk.PhotoImage(img11)
        
        # Down fourth-buttom
        b1= Button(bg_img, image=self.photoimg11, cursor='hand2',command=self.iExit)
        b1.place(x=1100, y=340, width=220, height=220)
        
        b1_1= Button(bg_img,text='Exit',cursor='hand2',command=self.iExit,font=('times new roman',14,'bold'),bg='darkblue',fg='white')
        b1_1.place(x=1100, y=520, width=220, height=40)
        
    def student_details(self):
        self.new_window=Toplevel(self.root)
        self.app=StudentSystem(self.new_window)
        
    def open_img(self):
        os.startfile("data")
        
    def train_data(self):
        self.new_window=Toplevel(self.root)
        self.app=Train(self.new_window)
        
    def face_data(self):
        self.new_window=Toplevel(self.root)
        self.app=FaceReconizerSystem(self.new_window)
    
    def attendance_data(self):
        self.new_window=Toplevel(self.root)
        self.app=AttendanceSystem(self.new_window)
        
    def developer_data(self):
        self.new_window=Toplevel(self.root)
        self.app=DeveloperSystem(self.new_window)
        
    def help_data(self):
        self.new_window=Toplevel(self.root)
        self.app=HelpSystem(self.new_window)
        
    def iExit(self):
        self.iExit=tkinter.messagebox.askyesno("Face Recognition","Are you Sure Exit this project",parent=self.root)
        if self.iExit>0:
            self.root.destroy()
        else:
            return
```

```python
if __name__=='__main__':
    root=Tk()
    obj = face_reconization_System(root)
    root.mainloop()
```
